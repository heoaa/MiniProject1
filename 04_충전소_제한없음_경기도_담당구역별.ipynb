```python
reg = ['안양시', '남양주시', '용인시', '김포시', '화성시', '부천시', '이천시', '성남시', 
       '여주시', '동두천시', '오산시', '의왕시', '평택시', '군포시', '하남시', '광명시']
```

```python
df_reg = df_g[(df_g['시군구'] == '안양시')|
              (df_g['시군구'] == '남양주시')|
              (df_g['시군구'] == '용인시')|
              (df_g['시군구'] == '김포시')|
              (df_g['시군구'] == '화성시')|
              (df_g['시군구'] == '부천시')|
              (df_g['시군구'] == '이천시')|
              (df_g['시군구'] == '성남시')|
              (df_g['시군구'] == '여주시')|
              (df_g['시군구'] == '동두천시')|
              (df_g['시군구'] == '오산시')|
              (df_g['시군구'] == '의왕시')|
              (df_g['시군구'] == '평택시')|
              (df_g['시군구'] == '군포시')|
              (df_g['시군구'] == '하남시')|
              (df_g['시군구'] == '광명시')]
```

```python
df_reg = df_reg.iloc[:,1:9]
```

## 집계

```python
pivot_reg_type = df_reg.pivot_table(values = '주소',
                             index = '시군구', columns='충전기타입',
                             aggfunc = 'count')
pivot_reg_type = pivot_reg_type.fillna(0)
pivot_reg_type
```

## 시각화

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rc('font', family = 'Malgun Gothic', size = 10)

plt.figure(figsize = (16,10))
sns.heatmap(pivot_reg_type, annot=True, fmt = '.0f', cmap='rocket_r')
plt.title('  ')
plt.show()
```
