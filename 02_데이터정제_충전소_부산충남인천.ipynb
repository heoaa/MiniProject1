```python
import pandas as pd
```

## 데이터 불러오기

```python
file_path = './data/rawdata/충전소_리스트.csv'
df_p = pd.read_csv(file_path)
df_p
```

```python
df_p.info()
```

## 결측치 처리

### 이용자 제한 NaN값 > 제한없음으로 변경

```python
df_p['이용자 제한'] = df_p['이용자 제한'].fillna('제한없음')
```

```python
df_p['지역'] = df_p['지역'].str.strip()
df_p.columns
```

```python
df_p['지역'].unique()
```

```python
df_p
```

## 이용자 제한이 있는 경우 제거

```python
df_n = df_p[df_p['이용자 제한'] == '제한없음']
```

```python
df_n
```

## 충남, 부산, 인천 데이터만 합치기

```python
d1 = df_n[df_n['지역'] == '충청남도']
```

```python
d2 = df_n[df_n['지역'] == '부산광역시']
```

```python
d3 = df_n[df_n['지역'] == '인천광역시']
```

```python
df=pd.DataFrame()
df = pd.concat([d1, d2], ignore_index=True)
df = pd.concat([df, d3], ignore_index=True)
df
```

## 중복 제거

```python
df[df.duplicated()]
```

```python
df[df.duplicated()].index
```

```python
df = df.drop(df[df.duplicated()].index, axis = 0)
```

```python
df
```

```python
df.reset_index(inplace=True, drop=True)
df
```
