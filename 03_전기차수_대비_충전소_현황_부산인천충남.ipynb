```python
import pandas as pd
```

## 전기차 대수 현황

```python
# rawdata
file_path1 = './1. 전기차/지역별_전기차_대수_현황.csv'

df_c = pd.read_csv(file_path1, skipfooter=37)
df_c
```

```python
df_car = df_c[['부산','인천','충남']]
df_car = df_car.rename(index={0:'전기차대수'})
df_car
```

```python
df_car = df_car.T
```

```python
df_car = df_car.rename(index={'부산':0, '인천':1, '충남':2})
df_car
```

```python
df_car['지역'] = ['부산광역시','인천광역시','충청남도']
df_car
```

## 충전소

```python
# newdata
file_path = './data/newdata/충전소_부산인천충남.csv'
df = pd.read_csv(file_path)
df
```

```python
df_sub = df.groupby('지역')['충전소'].count()
df_sub = df_sub.reset_index()
df_sub
```

```python
df = pd.merge(df_sub, df_car, on = '지역', how = 'inner')
df
```

```python
df['전기차/충전소'] = df['전기차대수']/df['충전소']
df
```

```python
# 팀원컬럼에 맞춰서 변경하기
df.columns = ['도시', '총 충전소', '총 전기차', '전기차/충전소']
df['도시'] = ['부산', '인천', '충남']
df = df[['도시','총 전기차', '총 충전소', '전기차/충전소']]
df
```

```python
df.to_csv('./data/newdata/1. 전기차/전기차수_대비_충전소_부산인천충남.csv', index=False)
```
