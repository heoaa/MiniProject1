```python
import folium
import json
```

```python
yongin_map = folium.Map(
                            location = [37.1778, 127.178],       # 최초에 보여줄 지도의 중앙점 좌표 [위도, 경도]
                            tiles = 'openstreetmap',                  # 지도 스타일 (openstreetmap : 일반적 도시 형태)
                            zoom_start = 13                           # 최초 보여줄 지도의 스케일
)
yongin_map
```

```python
# 데이터프레임에 있는 위도/경도 값을 각각 추출하여 출력

for i in range(0,len(df)):
    a = df['위도'][i]
    b = df['경도'][i]
    # print(a, b)
    
    # 지도에  표시할 마킹 모양 등 스타일 지정
    folium.CircleMarker(
        location = [a, b],         # 위치
        fill_color = 'yellow',      # 채우기 색상
        fill_opacity = 1,          # 채우기 불투명도
        color = 'green',          # 테두리 색상
        weight = 1,                # 테두리 굵기
        radius = 5                 # 전체 원의 지름
    ).add_to(yongin_map)
    
yongin_map
```
